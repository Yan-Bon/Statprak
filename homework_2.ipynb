```python
import matplotlib.pyplot as plt
import numpy as np


def beauty_derivative_plot(x0):
  x_t = np.linspace(x0 - 1, x0 + 1, 100)
  x = np.linspace(-2 * np.pi, 2 * np.pi, 100)

  f = lambda x: 2 * np.sin(x) - np.cos(2 * x)
  df = lambda x: 2 * np.cos(x) + 2 * np.sin(2 * x)
  y = f(x)
  dy = df(x)

  y0 = f(x0)
  k = df(x0)
  tangent = k * (x_t - x0) + y0


  plt.plot(x, y, label = 'f(x)')
  plt.plot(x, dy, label = "f'(x)")
  plt.plot(x_t, tangent, '--')

  plt.xlabel("x")
  plt.ylabel("y")

  plt.grid()

  plt.show()

beauty_derivative_plot(-3)
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.loadtxt('Копия populations.txt', delimiter='\t', skiprows=1)
years = data[:, 0]
hares = data[:, 1]
lynxes = data[:, 2]
carrots = data[:, 3]
```

```python
plt.plot(years, hares, label='Hares')
plt.plot(years, lynxes, label="Lynxes")
plt.plot(years, carrots, label="Carrots")

plt.legend()

plt.show()
```

```python
i_h = np.argmax(hares)
i_l = np.argmax(lynxes)
i_c = np.argmax(carrots)

year_h = years[i_h]
year_l = years[i_l]
year_c = years[i_c]

plt.plot(years, hares, label="Hares")
plt.plot(years, lynxes, label="Lynxes")
plt.plot(years, carrots, label="Carrots")

plt.scatter(year_h, hares[i_h])
plt.scatter(year_l, lynxes[i_l])
plt.scatter(year_c, carrots[i_c])

plt.xlabel("Year")
plt.ylabel("Population")
plt.legend()
plt.grid()

plt.show()
```

```python
total = hares + lynxes
hares_ratio = hares / total
lynxes_ratio = lynxes / total

plt.bar(years, hares_ratio, label="Hares")
plt.bar(years, lynxes_ratio, bottom=hares_ratio, label="Lynxes")

plt.xlabel("Year")
plt.ylabel("Ratio")
plt.legend()

plt.show()
```

```python
mean_h = np.mean(hares)
std_h = np.std(hares)

mean_l = np.mean(lynxes)
std_l = np.std(lynxes)

mean_c = np.mean(carrots)
std_c = np.std(carrots)

print("Hares:", mean_h, std_h)
print("Lynxes:", mean_l, std_l)
print("Carrots:", mean_c, std_c)

plt.figure(figsize=(15, 4))

plt.subplot(1, 3, 1)
plt.hist(hares)
plt.axvline(mean_h)
plt.title("Hares")

plt.subplot(1, 3, 2)
plt.hist(lynxes)
plt.axvline(mean_l)
plt.title("Lynxes")

plt.subplot(1, 3, 3)
plt.hist(carrots)
plt.axvline(mean_c)
plt.title("Carrots")

plt.show()
```

```python
result = ["H" if h > l else "L" for h, l in zip(hares, lynxes)]

print(result)
```

```python
hares_diff = np.diff(hares)
lynxes_diff = np.diff(lynxes)

years_diff = years[1:]

plt.plot(years_diff, hares_diff, label="Hares change")
plt.plot(years_diff, lynxes_diff, label="Lynxes change")

plt.xlabel("Year")
plt.ylabel("Change in population")
plt.legend()
plt.grid()

plt.show()

corr = np.corrcoef(hares_diff, lynxes_diff)[0, 1]
print("Correlation:", corr)
```

```python
plt.scatter(hares, lynxes)

plt.xlabel("Hares population")
plt.ylabel("Lynxes population")
plt.title("Hares vs Lynxes")

plt.grid()
plt.show()
```

```python
true_value = hares[-1]

hares_known = hares[:-1]

estimate = hares_known[-1] + (hares_known[-1] - hares_known[-2])

error_ratio = abs(estimate - true_value) / true_value

print("Estimated:", estimate)
print("True:", true_value)
print("Error ratio:", error_ratio)
```
